```python
import pandas as pd 
```

```python
df = pd.read_csv("owid-covid-data.csv")
df
```

```python
df.info()
```

```python
columns_needed = [
    "location",
    "continent",
    "date",
    "total_cases",
    "new_cases",
    "total_deaths",
    "new_deaths",
    "total_vaccinations",
    "population"
]

df = df[columns_needed]
df
```

```python
df = df.dropna(subset=["location", "date"])
df = df.dropna()
df = df.fillna(0)
df = df.drop_duplicates()
df
```

```python
df["date"] = pd.to_datetime(df["date"])
df
```

```python
df = df.sort_values(by=["location", "date"])
df
```

```python
df = df.sort_values(by=["location", "date"])

df = df[df["total_cases"] > 0]

df
```

```python
df = df[
    (df["new_cases"] > 0) |
    (df["new_deaths"] > 0) |
    (df["total_vaccinations"] > 0)
]
df
```

```python
df.to_excel("cleaned_covid_data.xlsx", index=False)
```
